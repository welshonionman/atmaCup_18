```python
import numpy as np
import pandas as pd
from src.constants import TRAIN_FEATURES, INTRINSIC_MATRIX

from src.viz.trajectory import (
    project_trajectory_to_image_coordinate_system,
    overlay_trajectory,
    get_trajectory,
    get_images,
)
from src.preprocess import add_id_scene_id_sec

np.set_printoptions(suppress=True)
```

```python
df_feature_train = pd.read_csv(TRAIN_FEATURES)
df_feature_train = add_id_scene_id_sec(df_feature_train)
df_feature_train = df_feature_train.sort_values(by=["id_scene", "id_sec"])
```

```python
idx = np.random.randint(0, len(df_feature_train))
idx += 1

id_ = df_feature_train.iloc[idx]["ID"]
trajectory = get_trajectory(df_feature_train, idx)
print(idx)
# print((trajectory * 100).round())
print((trajectory).round(2))
trajectory_image = project_trajectory_to_image_coordinate_system(
    trajectory, INTRINSIC_MATRIX
)
images = get_images(df_feature_train, idx)
overlay_trajectory(trajectory_image, images, id_)
```
